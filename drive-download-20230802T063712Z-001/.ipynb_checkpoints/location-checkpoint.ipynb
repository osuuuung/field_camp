```python
import os
from selenium import webdriver
import time
import pandas as pd
import numpy as np
from selenium.webdriver.common.by import By
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
east_data = pd.read_csv("C:\\Users\\ohesu\\Documents\\FIELD_CAMP\\drive-download-20230802T063712Z-001\\데이터\\데이터\\3. 멀어야하는 공공시설\\유흥시설_동구.csv", encoding='cp949')
middle_data = pd.read_csv("C:\\Users\\ohesu\\Documents\\FIELD_CAMP\\drive-download-20230802T063712Z-001\\데이터\\데이터\\3. 멀어야하는 공공시설\\유흥시설_중구.csv", encoding='cp949')
factory_data = pd.read_csv("C:\\Users\\ohesu\\Documents\\FIELD_CAMP\\drive-download-20230802T063712Z-001\\데이터\\데이터\\3. 멀어야하는 공공시설\\인천광역시_공장등록 현황_20230531.csv", encoding='cp949')
```

```python
# Geopy 초기화
geolocator = Nominatim(user_agent="geoapiExercises")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=2)
geolocator = Nominatim(user_agent="geoapiExercises", timeout=10)

# 위경도 정보 얻기
east_data['latitude'] = east_data['소재지(도로명)'].apply(lambda x: geocode(x).latitude if geocode(x) else None)
east_data['longitude'] = east_data['소재지(도로명)'].apply(lambda x: geocode(x).longitude if geocode(x) else None)

# 결과 저장
east_data.to_csv("유흥시설_동구_with_위경도.csv", encoding="cp949", index=False)
```

```python
# Geopy 초기화
geolocator = Nominatim(user_agent="geoapiExercises")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# 위경도 정보 얻기
middle_data['latitude'] = middle_data['소재지(도로명)'].apply(lambda x: geocode(x).latitude if geocode(x) else None)
middle_data['longitude'] = middle_data['소재지(도로명)'].apply(lambda x: geocode(x).longitude if geocode(x) else None)

# 결과 저장
middle_data.to_csv("유흥시설_중구_위경도.csv", encoding="cp949", index=False)
```
